# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt


table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT ConvertedComp 
FROM {}
""".format(table_name)

# Load data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Remove rows with null values in 'ConvertedComp'
df = df.dropna(subset=['ConvertedComp'])

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT Age 
FROM {}
""".format(table_name)

# Load data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Remove rows with null values in 'ConvertedComp'
df = df.dropna(subset=['Age'])

plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'])
plt.title('Box plot of Age')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM {}
""".format(table_name)

# Load data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Remove rows with null values in 'ConvertedComp'
df = df.dropna(subset=['Age','WorkWeekHrs'])

plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs 
FROM {}
""".format(table_name)

# Load data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Remove rows with null values in 'Age', 'WorkWeekHrs', or 'CodeRevHrs'
df = df.dropna(subset=['Age', 'WorkWeekHrs', 'CodeRevHrs'])

# Plotting the bubble plot
plt.figure(figsize=(10, 8))
bubble = plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*5, alpha=0.5, c=df['Age'])
plt.title('Bubble plot of WorkWeekHrs vs CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.colorbar(bubble, label='Age')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
table_name = 'DatabaseDesireNextYear'  # the table you wish to describe

QUERY = """
SELECT * 
FROM {}
Order by Respondent DESC
""".format(table_name)

# Load data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

df.head()


top_databases = df['DatabaseDesireNextYear'].value_counts().head(5)

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""

# Load data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Filtering data for the age group 30 to 35
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Grouping data by 'Age' and calculating median 'WorkWeekHrs' and 'CodeRevHrs'
median_data = filtered_df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

# Plotting the stacked chart
median_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.legend(title='Metrics')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT Age, ConvertedComp
FROM master
"""

# Load data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Filtering data for ages between 45 and 60
filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Grouping data by 'Age' and calculating median 'ConvertedComp'
median_comp = filtered_df.groupby('Age')['ConvertedComp'].median()

# Plotting the median ConvertedComp for ages between 45 to 60
plt.figure(figsize=(10, 6))
median_comp.plot(kind='bar', color='skyblue')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python

# Query to select 'MainBranch' column
QUERY = """
SELECT MainBranch
FROM master
"""

# Load data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Counting occurrences of each category in 'MainBranch'
branch_counts = df['MainBranch'].value_counts()

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
branch_counts.plot(kind='barh', color='skyblue')
plt.title('Count of MainBranch Categories')
plt.xlabel('Count')
plt.ylabel('MainBranch Categories')
plt.grid(axis='x')
plt.show()

```
